`range()` akan membuat **kumpulan angka** secara otomatis sesuai yang kita inginkan. Cara penggunaanya adalah `range(begin, end, step)`

````python
var_number = [10,22,33,40,50,60]

for x in var_number:
  if x%10 == 0 and x%2 == 0:
    print(f"{x} bilangan ini kelipatan 10 dan habis dibagi 2")
  elif x%10 != 0 and x%2 == 0:
    print(f"{x} bilangan ini bukan kelipatan 10 dan habis dibagi 2")
  else:
    print(f"{x} Gak tahu")
````

## 2.Tuple

Tuples are ordered collections of heterogeneous data that are unchangeable. Heterogeneous means tuple can store variables of all types.


Characteristic of Tuple :
1. Ordered
2. Heterogeneous item
3. Immutable
4. duplicate

**Why Tuples are Preferred over Lists?**

Tuples are preferred when the user **does not want** the data to be **modified**. Sometimes, the user can create an object that is intended to remain intact during its lifetime. Tuples are immutable, so they can be used to prevent accidental addition, modification, or removal of data.

Also, **tuples use less memory**, and they make program execution faster than using lists. Lists are slower than tuples because every time a new execution is done with lists, new objects are created, and the objects are not interpreted just once. Tuples are identified by Python as one immutable object. Hence, they are built as one single entity.

## 3.Sets

Sets are a collection of **unique heterogenous** and **unordered items**. Sets are mutable but can hold only unique values in the dataset. Set operations are similar to the ones used in arithmetic, including mathematical operations like union, intersection, difference etc.

Characteristic of Sets :
1. Unordered
2. Heterogeneous item
3. Mutable Item
4. Unique
5. Set Operations : Union, Intersection,Difference, Symmetric Difference


One of the ways that sets are used is when checking whether or not some elements are contained in a set or not. For example, sets are highly optimized for membership tests. They can be used to check whether a set is a subset of another set and to identify the relationship between two sets.

### 3.1 Create sets

````python
my_set = {1, 2, 3, 4, 5, 5, 5} #create set
print(my_set)
````

````python
book_set = set(("Harry Potter", "Angels and Demons", "Atlas Shrugged"))
print(book_set)
````

````python
# create set from list
number_list = [20, 30, 20, 30, 50, 30]
# create a set from a list
sample_set = set(number_list)

sample_set
````

### 3.1 Add item using sets

Though the value of the item in a Set can’t be modified. We can add new items to the set using the following two ways.

1. The `add()` method: The `add()` method is used to add one item to the set.
2. Using `update()` Method: The `update()` method is used to multiple items to the Set. We need to pass the list of items to the update() method

````python
#add  method to add one item
book_set.add('The God of Small Things')
print(f"Menggunakan metode add {book_set}")

# update() method to add more than one item
book_set.update(['Atlas Shrugged', 'Ulysses'])
print(f"Menggunakan metode update {book_set}")
````

### 3.2  Removing sets

- `remove` : To remove a single item from a set. This method will take one parameter, which is the item to be removed from the set. Throws a keyerror if an item not present in the original set

- `discard` : To remove a single item that may or may not be present in the set. This method also takes one parameter, which is the item to be removed. If that item is present, it will remove it. It will not throw any error if it is not present.

- `pop` : To remove any random item from a set

````python
# remove single item
book_set.remove('The God of Small Things')
print(f"metode remove {book_set}")

# remove any random item from a set
book_set.pop()
print(f"metode pop {book_set}")
````

### 5.2 Map Functions

`map()` function returns a map object(which is an iterator) of the results after applying the given function to each item of a given iterable (list, tuple etc.)

map(function, iter)

- function : It is a function to which map passes each element of given iterable.
- iter : It is a iterable which is to be mapped.

````python
# Return double of n
def squared(n):
    return n * n

# We double all numbers using map()
numbers = (1, 2, 3, 4)
result = map(squared, numbers)
print(list(result))
````

````python
#using lamda expression

numbers = (1, 2, 3, 4)
result = map(lambda x: x * x, numbers)
print(list(result))
````
