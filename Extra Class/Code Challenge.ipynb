3. Buatlah program Python yang meminta pengguna untuk memasukkan sebuah bilangan bulat positif, lalu menghitung dan mencetak faktorial dari bilangan tersebut. Faktorial dari bilangan bulat n (dilambangkan dengan n!) adalah hasil perkalian semua bilangan bulat positif dari 1 hingga n.

  Example : <br>
Masukkan bilangan bulat positif: 5
Faktorial dari 5 adalah: 120


````python
def menghitung_faktorial(n):
    if n < 0:
        return "Faktorial tidak didefinisikan untuk bilangan negatif"
    elif n == 0 or n == 1:
        return 1
    else:
        faktorial = 1
        for i in range(1, n + 1):
            faktorial *= i
        return faktorial
input_n = int(input('Masukan bilangan positif : '))
hasil_faktorial = menghitung_faktorial(input_n)
print("Faktorial dari", input_n , 'adalah' , hasil_faktorial)

````

  Masukkan bilangan bulat positif: 12
12 bukan bilangan prima.

5. Buatlah program Python yang meminta pengguna untuk memasukkan sebuah kata atau frase, dan kemudian menentukan apakah kata atau frase tersebut merupakan palindrom (bisa dibaca sama dari depan maupun dari belakang).

  Masukkan sebuah kata atau frase: radar
radar adalah palindrom.

  Masukkan sebuah kata atau frase: hello
hello bukan palindrom.

````python
def cek_palindrom(kata):
  cleaned=kata.replace(' ','').lower()
  return cleaned==cleaned[::-1]
masukan=input("Masukan kata: ")
if cek_palindrom(masukan):
  print(f'{masukan} adalah palindrom')
else:
  print(f'{masukan} bukan palindrom')
````

Pak maman ingin menukarkan uangnya sejumlah RP. 32.892.700 kedalam pecahan uang Rp. 100.000 Rp 50.000 Rp. 10.000 Rp 5.000 Rp 1000 Rp 500 dan Rp 100 ke pak budi, Buatlah pemrograman python yang bisa membantu mempercepat pak Budi dalam menentukan berapa buah/lembar uang untuk masih masih pecahan tersebut

````python
def tukar_pecahan(uang):
    pecahan = [100000, 50000, 10000, 5000, 1000, 500, 100]
    hasil = {}

    for nominal in pecahan:
        jumlah = uang // nominal
        uang %= nominal
        hasil[nominal] = jumlah
    return hasil
tukar=32892700
hasil_tukar=tukar_pecahan(tukar)

print(f"Jumlah uang yang ditukar: Rp. {tukar:,}")
print("Rincian pecahan:")
for nominal in hasil_tukar:
    jumlah = hasil_tukar[nominal]
    print(f"Rp. {nominal:,}   : {jumlah} lembar")
````

Namun, karena kasirnya masih menggunakan sistem manual, maka semakin lama kasir mengalami kerepotan dalam menangani transaksi perharinya. Input program adalah total belanja pelanggan, sedangkan outputnya adalah total belanja setelah pembulatan ke bawah ke kelipatan Rp 1000

11. Hitunglah Upah Kerja (Kotor) dari seluruh karyawan Perusahaan B jika diketahui upah per jam adalah Rp 30000

  Tampilkan output sebagai berikut

  Contoh:

  - Upah per jam = 30000
  - Upah Kerja (Kotor) = Jam Kerja x Upah per Jam

Upah Kerja (Kotor) Chandra = Rp 1500000
Upah Kerja (Kotor) Hartono = Rp 1260000

````python
def upah_kotor(jam, upah_perjam):
  return jam*upah_perjam
karyawan = {
    'Doni' : 10,
    'Joko' : 15,
    'Alex' : 8,
    'Chandra' : 13,
    'Hartono' : 11,
    'Samsul' : 9
}
upah_perjam=30000
for nama, jam in karyawan.items():
  upah=upah_kotor(jam,upah_perjam)
  print(f"Upah kerja (kotor) {nama} = Rp {upah:,}")
````
