## Dibimbing Day 9 -- Hypothesis Testing Concept

```python
import scipy.stats as stats
import pandas as pd
import numpy as np
```

#### Generate Data

```python
time = np.array([18, 160, 234, 149, 145, 107, 197, 75, 201, 225, 211, 119,
              157, 145, 107, 244, 163, 114, 145,  68, 111, 185, 202, 146,
              203, 224, 213, 104, 178, 166, 187, 154, 177, 95, 185, 50, 110,
              216, 138, 151, 166, 135, 155, 84, 248, 173, 131, 207, 121, 135])
```

#### Problem 1

Is the average completion time of the final test 175 minutes? We have average test times spent in the final exam.

Confident interval is %95 so alpha is (0.05)

First step set up the hypothesis

H0:  = 175

H1:   175

#### Data Description

```python
stats.describe(time)
```

#### Hypothesis Test - Calculating P-Value

```python
stats.ttest_1samp(time, popmean =175) #popmean = acuan rata rata (h0)
```

H0:  = 175

H1:   175


if P-value > alpha(0.05) we can't deny the HO hypothesis

if P-value < alpha(0.05) we can deny the HO hypothesis

n this case our P-value is 0.007 and its less than alpha(0.05) we can deny the HO hypothesis.

### Conclusion: Hypothesis test proves that the mean or average of the completion time in final test is not equal to 175 minutes.

#### Problem 2

Is the average completion time of the final test is larger than 150 minutes? We have average test times spent in the final exam.

Confident interval is %95 so alpha is (0.05)

First step set up the hypothesis

H0:  = 150

H1:  > 150

#### Hypothesis Test - Calculating P-Value

```python
stats.ttest_1samp(time, popmean =150, alternative = "greater") #one tail, greater lebih besar less yg lebih kecil
```

H0:  = 150

H1:  > 150


if P-value > alpha(0.05) we can't deny the HO hypothesis

if P-value < alpha(0.05) we can deny the HO hypothesis

n this case our P-value is 0.26 and its larger than alpha(0.05), we can't deny the HO hypothesis.

### Conclusion: Hypothesis test does not prove that the mean is larger than 150.
